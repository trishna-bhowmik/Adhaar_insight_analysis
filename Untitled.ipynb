```python
#Importing Required Libraries:
import pandas as pd
import matplotlib.pyplot as plt

# Optional: better visuals
plt.style.use("default")
```

```python
#Loading One Aadhaar CSV File:
df=pd.read_csv("api_data_aadhar_enrolment_500000_1000000.csv")
df.head()
```

```python
df.columns
```

```python
#Creating Total Enrolments:
df['total_enrolments']=(
    df['age_0_5']+
    df['age_5_17']+
    df['age_18_greater']
)
df.head()
```

```python
#Creating Age Group Graph:
age_totals = df[['age_0_5', 'age_5_17', 'age_18_greater']].sum()
age_totals.plot(kind='bar')
plt.title("Aadhaar Enrolments by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Enrolments")
plt.show()
```

# Observation:

The 18+ age group has the lowest number of enrolments.  
The 5–17 age group has moderate enrolments.  
The 0–5 age group has the highest enrolments.

```python
#Creating Top States In Aadhaar Enrolments:
state_data=df.groupby('state')['total_enrolments'].sum()
state_data=state_data.sort_values(ascending=False)
state_data.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 States by Aadhaar Enrolments")
plt.xlabel("State")
plt.ylabel("Total Enrolments")
plt.show()
```

# Observation:

Uttar Pradesh dominate Aadhaar enrolments.

```python
#Creating Daily Aadhaar Enrolments:
df['date']=pd.to_datetime(df['date'], format='%d-%m-%Y')
daily_enrolments=df.groupby('date')['total_enrolments'].sum()

plt.figure(figsize=(12,5))
plt.plot(daily_enrolments)
plt.title("Daily Aadhaar Enrolments Over Time")
plt.xlabel("Date")
plt.ylabel("Total Enrolments")
plt.show()
```

```python
#Creating Age Group Trend Over Time:
age_time=df.groupby('date')[['age_0_5','age_5_17','age_18_greater']].sum()

plt.figure(figsize=(12,5))
plt.plot(age_time['age_0_5'], label='Age 0-5')
plt.plot(age_time['age_5_17'], label='Age 5-17')
plt.plot(age_time['age_18_greater'], label='Age 18+')

plt.legend()
plt.title("Age-wise Aadhaar Enrolment Trend")
plt.xlabel("Date")
plt.ylabel("Enrolments")
plt.show()
```

# Observation:

Enrolments for 18+ age group remained consistent over time.

```python
#Creating Cumulative Enrolments:
cumulative=daily_enrolments.cumsum()

plt.figure(figsize=(12,5))
plt.plot(cumulative)
plt.title("Cumulative Aadhaar Enrolments (Growth Curve)")
plt.xlabel("Date")
plt.ylabel("Cumulative Enrolments")
plt.show()
```

# Observation:

The curve continuously increases and never decreases.

```python
#Creating Distribution of Enrolments:
plt.figure(figsize=(10,5))
plt.hist(df['total_enrolments'], bins=50)
plt.title("Distribution of Aadhaar Enrolments per Record")
plt.xlabel("Enrolments")
plt.ylabel("Frequency")
plt.show()
```

```python
#Creating Age Group Share:
age_totals=df[['age_0_5','age_5_17','age_18_greater']].sum()

plt.figure(figsize=(6,6))
plt.pie(age_totals, labels=age_totals.index, autopct='%1.1f%%')
plt.title("Age Group Contribution to Aadhaar Enrolments")
plt.show()
```

# Observation:

# Age Group Bar Graph vs Pie Chart
Both graphs confirm that 18+ age group dominates enrolments.  
The bar graph shows absolute numbers while the pie chart shows percentage contribution.

```python
#Creating State-wise Accumulation:
top_states=df.groupby('state')['total_enrolments'].sum().sort_values(ascending=False).head(5).index

state_time=df[df['state'].isin(top_states)]
state_time=state_time.groupby(['date','state'])['total_enrolments'].sum().unstack()

state_time.plot(kind='area', figsize=(12,5))
plt.title("Top 5 States Aadhaar Enrolment Over Time")
plt.xlabel("Date")
plt.ylabel("Enrolments")
plt.show()
```
